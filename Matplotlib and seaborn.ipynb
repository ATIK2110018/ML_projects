# Matplotlib & Seaborn

```python
import numpy as np
import pandas as pd 
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df=pd.read_csv("./data/gapminder.csv")
```

```python
df.head()
```

```python
# Histogram
plt.hist(df.life_exp)
```

```python
#bin size , default is 10
plt.hist(df.life_exp,bins=15)
```

```python
# adding color
plt.hist(df.life_exp, bins=15 , color= "green")
```

```python
plt.hist(df.life_exp, bins=15 , color= "green")
plt.xlabel("life expectation")
plt.ylabel("frequency")
plt.title("distribution of expectation of life",fontsize=20)

#adjust layout for better visualization
plt.tight_layout()

#display the plot
plt.show
```

```python
#customizing figure size
```

```python
plt.figure(figsize=(10,6))
plt.hist(df.life_exp, bins=15 , color= "green")
plt.xlabel("life expectation")
plt.ylabel("frequency")
plt.title("distribution of expectation of life",fontsize=20)

#adjust layout for better visualization
plt.tight_layout()

#display the plot
plt.show
```

```python
#plot style
plt.style.available
```

```python
plt.style.use('fivethirtyeight')
plt.figure(figsize=(10,6))
plt.hist(df.life_exp, bins=15 , color= "green")
plt.xlabel("life expectation")
plt.ylabel("frequency")
plt.title("distribution of expectation of life",fontsize=20)

#adjust layout for better visualization
plt.tight_layout()

#display the plot
plt.show
```

```python
plt.style.use('fivethirtyeight')
plt.figure(figsize=(5,3))
plt.hist(df.life_exp, bins=15 , color= "green")
plt.xlabel("life expectation")
plt.ylabel("frequency")
plt.title("distribution of expectation of life",fontsize=20)

#adjust layout for better visualization
plt.tight_layout()

#display the plot
plt.show
#save figure
plt.savefig("distribution.pdf")
```

## Seaborn

### Visualization of statistical relationship

```python
df=pd.read_csv("./data/pulse.csv")
df.head()
```

```python
sns.relplot(data= df , x="age" , y="height")
```

```python
#categorizing by sex
sns.relplot(data= df , x="age" , y="height", hue = "sex")
```

```python
sns.relplot(data= df , x="age" , y="height", hue="sex", style="sex")
#it will change the mark acording to sex
```

```python
# we can style it according to another data too
sns.relplot(data= df , x="age" , y="height", hue="sex", style="night_id")
```

## heatmap

```python
df=pd.read_csv("./data/pulse1.csv")
df.head()
```

```python
df.corr()
```

```python
plt.figure(figsize=(6,4))
sns.heatmap(df.corr())
```

```python
plt.figure(figsize=(20,5))
sns.heatmap(df.corr(), annot=True)
```

```python
## here we can see who is correlated with whom , and its value
```

```python
## changing color
plt.figure(figsize=(20,5))
sns.heatmap(df.corr(), annot=True,cmap="Greens")
```

```python
## one function to do all 
sns.pairplot(df)
```
